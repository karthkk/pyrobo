```python
import numpy as np
import cPickle
from matplotlib import pyplot as pl
```

```python
%pylab inline
```

```python
import pylibcam
```

```python
c1 = pylibcam.PyCamera('/dev/video1', 640, 480, 5)
c2 = pylibcam.PyCamera('/dev/video2', 640, 480, 5)
```

```python
c1.update(25, 2000)
c2.update(25, 2000)
```

```python
i1 = c1.toRGB()
pl.imshow(i1)
```

```python
i2 = c2.toRGB()
pl.imshow(i2)
```

```python
c1.StopCam()
c2.StopCam()
```

```python
c1.setExposure(1)
c2.setExposure(1)
```

```python
c1.setBrightness(1)
c2.setBrightness(1)
```

```python
import cv2
```

```python
g1 = cv2.cvtColor(i1,cv2.COLOR_RGB2GRAY)
```

```python
g2 = cv2.cvtColor(i2,cv2.COLOR_RGB2GRAY)
```

```python
pl.imsave('/tmp/g1.png', g1)
```

```python
pl.imsave('/tmp/g2.png', g2)
```

```python
cv2.imwrite('/tmp/g1.pgm', g1)
cv2.imwrite('/tmp/g2.pgm', g2)
```
