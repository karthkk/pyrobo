```python
import pylibcam
```

```python
cam_1 = pylibcam.PyCamera('/dev/video0', 640, 480, 5)
cam_2 = pylibcam.PyCamera('/dev/video1', 640, 480, 5)
```

```python
import time
cam_1.setBrightness(1)
cam_2.setBrightness(1)
for i in range(4):
    time.sleep(2)
    cam_1.update(25,1000)
    cam_2.update(25,1000)
cam_1.update(25,1000)
cam_2.update(25,1000)
img_1 = cam_1.toRGB()
img_2 = cam_2.toRGB()
```

```python
img = np.zeros((480, 640*2, 3), dtype=img_1.dtype)
img[:,:640,:] = img_1
img[:,640:,:] = img_2
pl.imshow(img)
```
